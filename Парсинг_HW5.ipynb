```python
from pymongo import MongoClient
```

```python
client = MongoClient()
```

```python
print(client.list_database_names())
```

```python
db = client.skazki_labirint
print(db.list_collection_names())
```

```python
list_books = []
for collections in db.list_collection_names():
  for book in list(db[collections].find()):
    list_books.append(book)
```

```python
print(f'всего спарсили {len(list_books)} книг') 
```
