### 2. Работа будет состоять с недокументированным API. Нужно ввести релевантный запрос на сайте https://www.delivery-club.ru/search

##### (а) из предложенных точек с помощью API найти долю (в %) с бесплатной и платной доставкой. Для каждой категории рассчитать среднюю минимальную стоимость заказа.  


```python
import requests
from pprint  import pprint
import pandas as pd
import numpy as np
```

```python
params = {
    'latitude': '58.520437',
    'longitude': '31.264759',
    'query': 'суши'
}

url = 'https://api.delivery-club.ru/api1.2/vendors/search'

response = requests.get(url = url, params =params)
print(response.status_code)
pprint(response.json())
```

```python
# посмотрим названия точек, педлагающих суши 
vendors = response.json()['vendors']
for vendor in vendors:
    print(vendor['name'])
```

```python
# посмотрим названия и категории точек, педлагающих суши 
vendors = response.json()['vendors']
for vendor in vendors:
    print(vendor['name'], vendor['categoryId'])
```

```python
restaurant=[]
shop=[]
for i in vendors:
    if i['categoryId']==1:
        restaurant.append(i['name'])
    else:    
        shop.append(i['name'])

print('Рестораны: \n\n', restaurant, '\n\nМагазины: \n\n',shop)
```

```python
# посмотрим названия и предложения точек, педлагающих суши 
vendors = response.json()['vendors']
for vendor in vendors:
    print(vendor['name'], vendor['delivery'])
```

```python
# разбиваем точки с бесплатной и платной доставкой 
pay_price_delivery, free_price_delivery, vendors = {}, {}, response.json()['vendors']
for i in vendors:
    if i['delivery']['price']['value'] == 0:
        free_price_delivery[i['name']] = [i['delivery']['minOrderPrice']['value'], i['delivery']['price']['value'], i['categoryId']]
    else:
        pay_price_delivery[i['name']] = [i['delivery']['minOrderPrice']['value'], i['delivery']['price']['value'], i['categoryId']]
```

```python
pprint(pay_price_delivery)
```

```python
pprint(free_price_delivery)
```

```python
# находим долю каждой доставки 
x=(len(free_price_delivery)*100)/(len(free_price_delivery)+len(pay_price_delivery))

print(len(free_price_delivery))
print(len(pay_price_delivery))
print(f'доля платной доставки: {int(x)}%,\nдоля бесплатной доставки: {100-int(x)}%')
```

```python
# средняя цена минимального заказа для каждой категории точек
all_min_oder_price = []
free_min_oder_price = []
pay_min_oder_price = []
for i in vendors:
    all_min_oder_price.append(i['delivery']['minOrderPrice']['value'])
    if i['delivery']['price']['value'] == 0:
        free_min_oder_price.append(i['delivery']['minOrderPrice']['value'])
    else:
        pay_min_oder_price.append(i['delivery']['minOrderPrice']['value'])      

print(f'средняя стоимость заказа по всем точкам составляет {round(np.mean(all_min_oder_price), 2)} рубл.')
print(f'средняя стоимость заказа по точкам с бесплатной доставкой составляет {round(np.mean(free_min_oder_price), 2)} рубл.')
print(f'средняя стоимость заказа по точкам с платной доставкой составляет {round(np.mean(pay_min_oder_price), 2)} рубл.')
```

#####  (б) для каждой из категорий из пункта (а) рассчитать долю (в %) магазинов и ресторанов

соберём сформированные словари в датафреймы

```python
# с платной доставкой
df_pay_price_delivery = pd.DataFrame(data=pay_price_delivery, index=['order_price', 'delivery_price', 'category']).T
#df_pay_price_delivery
```

```python
# с бесплатной доставкой
df_free_price_delivery = pd.DataFrame(data=free_price_delivery, index=['order_price', 'delivery_price', 'category']).T
#df_free_price_delivery
```

```python
# объединим их
general_df = pd.concat([df_free_price_delivery, df_pay_price_delivery], axis=0)
general_df
```

```python
restaurant_free = general_df[(general_df.category == 1) & (general_df.delivery_price == 0)].shape[0]
print(f'количество ресторанов с бесплатной доставкой {restaurant_free},  доля  {np.round(restaurant_free/general_df.shape[0]*100, 2)}%')
restaurant_pay = general_df[(general_df.category == 1) & (general_df.delivery_price > 0)].shape[0]
print(f'количество ресторанов с платной доставкой  {restaurant_pay}, доля  {np.round(restaurant_pay/general_df.shape[0]*100, 2)}%')
shop_free = general_df[(general_df.category == 5) & (general_df.delivery_price == 0)].shape[0]
print(f'количество магазинов с бесплатной доставкой  {shop_free}, доля {np.round(shop_free/general_df.shape[0]*100, 2)}%')
shop_pay = general_df[(general_df.category == 5) & (general_df.delivery_price > 0)].shape[0]
print(f'количество магазинов с платной доставкой {shop_pay}, доля {np.round(shop_pay/general_df.shape[0]*100, 2)}%')
```
